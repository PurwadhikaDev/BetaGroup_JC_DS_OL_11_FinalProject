## **Context**  

Term deposit, juga dikenal sebagai deposito berjangka adalah produk keuangan yang ditawarkan oleh bank dan lembaga keuangan lainnya kepada nasabah mereka. Dalam term deposit, nasabah setuju untuk menyetor sejumlah uang tertentu ke dalam rekening khusus untuk jangka waktu yang telah ditentukan, yang bisa berkisar dari beberapa bulan hingga beberapa tahun. Selama periode ini, dana nasabah menjadi tidak dapat dicairkan atau hanya bisa dicairkan dengan biaya tambahan yang biasa disebut denda. Term deposit menguntungkan bagi bank karena mereka dapat menggunakan dana yang diterima dari nasabah dalam bentuk deposito ini untuk untuk diinvestasikan uangnya pada produk keuangan lain yang memberikan rate of return (RoR) yang lebih tinggi daripada yang dibayarkan bank kepada nasabah atas penggunaan dananya. Selain itu, deposito memberikan stabilitas dalam sumber dana bank, membantu mereka memenuhi persyaratan permodalan yang diatur oleh otoritas pengawas, dan mendukung pertumbuhan bisnis mereka. Dalam marketing term deposit, bagaimana menyeleksi dan mengidentifikasi nasabah yang paling sesuai dan berpotensi untuk produk tersebut. Tujuan dari marketing ini adalah untuk menciptakan campaign marketing yang efisien dan efektif, yang mengarah pada peningkatan konversi nasabah baru dalam produk term deposit.  Dengan mengefektifkan seleksi calon nasabah potensial, bank dapat menghemat waktu, sumber daya, dan biaya marketing sambil meningkatkan kemungkinan penarikan nasabah baru dan pertumbuhan bisnis mereka.

## **Metric Evaluation**

- True Positive: Model memprediksi nasabah membuka deposito berjangka dan kenyataannya nasabah membuka deposito berjangka
- False Positive: Model memprediksi nasabah membuka deposito berjangka namun kenyataannya nasabah tidak membuka deposito berjangka
- False Negative: Model memprediksi nasabah tidak membuka deposito berjangka namun kenyataannya nasabah membuka deposito berjangka
- True Negative: Model memprediksi nasabah tidak membuka deposito berjangka dan kenyataannya nasabah tidak membuka deposito berjangka

Error type 1 (False Positive):
Konsekuensi: Biaya pemasaran dan sumber daya klien (Bank Portugal) terbuang percuma.

Error type 2 (False Negative):
Konsekuensi: Kehilangan nasabah potensial.

kami akan membuat asumsi tentang dampak kerugian dari kedua tipe error diatas adalah berikut ini.

- Biaya rata-rata menelepon di portugal adalah `$0.49/menit` berdasarkan sumber ini (https://www.onesimcard.com/cell-phone-rates/portugal/#rates_table)
- Gaji rata-rata di Portugal pada tahun 2008-2010 adalah `$21,664/tahun` berdasarkan sumber ini (https://www.statista.com/statistics/419498/average-annual-wages-portugal-y-on-y/)
- Deposit Interest Rate di Portugal pada tahun 2008-2010 paling tinggi ada di angka `4%` berdasarkan sumber ini (https://tradingeconomics.com/portugal/interest-rate)
- Lending Interest Rate di Portugal pada tahun 2008-2010 paling tinggi ada di angka `8%` berdasarkan sumber ini (https://tradingeconomics.com/portugal/bank-lending-rate)
- Biaya rata-rata living cost di portugal adalah `$11,665.92/tahun` berdasarkan sumber ini (https://www.numbeo.com/cost-of-living/historical-data?itemId=105&city_id=6029&name_city_id=Lisbon%2C+Portugal)

Biaya False Positive:
- Jika berdasarkan dataset, rata-rata duration ada di `258.2 detik` dan rata-rata bank menelepon calon nasabah dalam 1 campaign adalah `3 (2.56) kali`. maka biaya False Positive setiap 1 orang adalah (258.2 detik / 60) * ($0.49/menit) * 3 = `$6.41`

Biaya False Negative:
- Jika gaji dikurangi biaya hidup di portugal adalah ($21,664 - $11,665.92) = `$9,998` dan jika kita asumsikan uang tersebut `50%` nya untuk ditaruh di deposito berjangka maka setiap nasabah dapat membuka deposito berjangka dengan jumlah uang `$4,999`. jika kita hitung bunga yang didapatkan selama setahun jika nasabah membuka deposito berjangka adalah ($4,999 * (8% - 4%)) = `$199.96`

Berdasarkan konsekuensi dari setiap jenis kesalahan, fokus utama dari model yang akan kita buat adalah meminimalkan Error type 2 karena konsekuensi dari False Negative lebih serius daripada False Positives (Type I error). Error tipe 2 ini berpotensi menurunkan pendapatan klien (Bank Portugal) yang diakibatkan dari kehilangan nasabah potensial yang akan membuka deposito berjangka. Jadi nanti metric utama yang akan kita gunakan adalah `recall score`.

```python
# Inisialisasi dictionary untuk menyimpan nama kolom dan jumlah "unknown"
unknown_counts = {}

# Loop melalui setiap kolom dan hitung jumlah "unknown"
for column in df.columns:
    unknown_count = df[column].eq('unknown').sum()
    unknown_counts[column] = unknown_count
```

```python
unknown = {}
unknown_column = ['job','marital', 'education', 'housing','loan','default']
for kolom in unknown_column:
    unknown[kolom] = float(df[df[kolom] == 'unknown'].groupby(kolom).count()['y'].values/df.shape[0] * 100)
pd.DataFrame(unknown, index=['Proporsi Unknown']).T
```

missing value berdasarkan tabel diatas adalah berikut ini:
- `job`: terdapat 0.8% missing value dari kolom job
- `marital`: terdapat 0.19% missing value dari kolom marital
- `education`: terdapat 4.2% missing value dari kolom education
- `default`: terdapat 20.87% missing value dari kolom default
- `housing`: terdapat 2.4% missing value dari kolom housing
- `loan`: terdapat 2.4% missing value dari kolom loan

Dengan informasi ini, kita bisa mempertimbangkan beberapa strategi untuk menangani nilai "unknown" di tiap kolom:

- Kolom job, marital, dan education:

Menggantikan "unknown" dengan nilai modus mungkin merupakan pendekatan yang baik. Ini karena jumlah missing values relatif kecil dibandingkan dengan ukuran dataset.

- Kolom default:

Mengingat jumlah missing values yang sangat tinggi, menggantinya dengan modus ("no") mungkin tidak menggambarkan data dengan akurat. Mungkin lebih baik untuk menjadikan "unknown" sebagai kategori tersendiri atau mempertimbangkan untuk menghapus kolom ini jika analisis menunjukkan bahwa ia memiliki pengaruh kecil terhadap variabel target.

- Kolom housing dan loan:

### Outliers

Kita akan menggunakan metode statistik untuk mengidentifikasi outliers. Salah satu metode yang umum digunakan adalah IQR (Interquartile Range).

Dengan metode IQR, kita dapat menghitung batas bawah dan batas atas untuk setiap kolom dan kemudian menentukan data mana yang dianggap sebagai outlier.

```python
df_copy_numerik = df_copy.select_dtypes(exclude='object')
```

```python
iqr_data = {}

for col in df_copy_numerik:
    Q1 = df_copy[col].quantile(0.25)
    Q3 = df_copy[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    iqr_data[col] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound
    }

iqr_df = pd.DataFrame(iqr_data).T
iqr_df
```

Jika suatu nilai dalam kolom berada di bawah "lower_bound" atau di atas "upper_bound", maka nilai tersebut dapat dianggap sebagai outlier.

selanjutkan kita akan liat berapa banyak outliers yang ada pada setiap kolom

buat perbandingan modeling dengan outliers dan tanpa outliers

```python
def detect_outliers(df_copy, column):
    Q1 = df_copy[column].quantile(0.25)
    Q3 = df_copy[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df_copy[column] < lower_bound) | (df_copy[column] > upper_bound)].shape[0]
```

## Modeling

```python
logreg = LogisticRegression(solver='liblinear')
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier(random_state=2023)
rf = RandomForestClassifier(random_state=2023)
abc = AdaBoostClassifier(random_state=2023)
xgb = XGBClassifier(random_state=2023)
lgbm = lgb.LGBMClassifier(random_state=2023)
gbc = GradientBoostingClassifier(random_state=2023)

scaler = RobustScaler()

skfold = StratifiedKFold(n_splits=5)
```

Alasan penggunaan Robust Scaler dipilih karena teknik ini memanfaatkan informasi statistik seperti nilai median dan Interquartile Range (IQR) untuk menyesuaikan skala data. Kelebihan Robust Scaler terletak pada kemampuannya menangani dataset yang tidak terdistribusi normal atau mengandung pencilan (outliers).

Dalam proses ini, nilai median digunakan sebagai pusat skala, yang membuat Robust Scaler lebih tahan terhadap pengaruh nilai ekstrem atau pencilan dalam dataset. Selain itu, IQR digunakan untuk mengukur sebaran data, sehingga transformasi skala yang diterapkan menjadi lebih stabil dan efektif.

Stratified k-fold adalah suatu metode dalam validasi silang (cross-validation) di mana data dibagi menjadi k fold dengan mempertahankan proporsi distribusi kelas yang seimbang di setiap lipatan (fold). Pendekatan ini sangat berguna ketika kita memiliki kelas target yang tidak seimbang dalam dataset, yang berarti jumlah sampel yang termasuk dalam setiap kelas tidak proporsional.

```python
models = [logreg,knn,dt,rf,abc,xgb,lgbm,gbc]
list_model = []
score = []
std = []

for i in models:
    estimator=Pipeline([
        ('preprocess',transformer),
        ('model',i)])
    model_cv=cross_val_score(estimator,X_train_val,y_train_val,cv=skfold,scoring = 'recall')

    list_model.append(i)
    score.append(model_cv.mean())
    std.append(model_cv.std())

pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest','AdaBoost', 'XGBoost', 'LightGBM', 'Gradient Boosting'],'recall_score':score,'sdev':std}).set_index('model').sort_values(by='recall_score',ascending=False)
```

jika kita lihat modeling tanpa scaling, hasil recall score dari model terbaik masih berada di rentang `0.327046` dengan standar deviasi di angka `0.019875`. selanjutnya kita akan coba scaling menggunakan robust scaler.

### Modeling Dengan Scaling

### Model Benchmarking: menggunakan resampling

```python
samplers = [SMOTE(random_state=2023), NearMiss(), NearMiss(version=2), NearMiss(version=3), RandomOverSampler(random_state=2023),
            RandomUnderSampler(random_state=2023), SMOTEENN(random_state=2023)]
```

Ketujuh metode resampling ini adalah teknik yang akan digunakan digunakan untuk mengatasi masalah imbalance kelas dalam dataset. Berikut adalah penjelasan singkat untuk setiap metode:

1. SMOTE (Synthetic Minority Over-sampling Technique)
- SMOTE bekerja dengan membuat sampel sintetis baru di antara titik data minoritas yang sudah ada. Ini dilakukan dengan memilih titik acak dari kelas minoritas dan menggabungkannya dengan tetangga terdekatnya, menciptakan sampel baru di sepanjang garis yang menghubungkan titik tersebut.

2. NearMiss (version=1):
- NearMiss versi 1 memilih sampel mayoritas yang memiliki jarak minimum ke tiga sampel terdekat dari kelas minoritas.
Strategi ini bertujuan untuk meningkatkan presisi dengan memilih sampel mayoritas yang sangat dekat dengan kelas minoritas.

3. NearMiss (version=2):
- NearMiss versi 2 memilih sampel mayoritas yang memiliki jarak rata-rata terkecil ke tiga sampel terdekat dari kelas minoritas.
Strategi ini berfokus pada mempertahankan variasi yang lebih baik dengan memilih sampel mayoritas yang mewakili rata-rata jarak keseluruhan.

4. NearMiss (version=3):
- NearMiss versi 3 memilih sampel mayoritas yang memiliki jarak maksimum ke tiga sampel terdekat dari kelas minoritas.
Strategi ini bertujuan untuk mengurangi redundansi dalam sampel mayoritas dengan memilih sampel yang paling berbeda dari kelas minoritas.

5. RandomOverSampler
- RandomOverSampler bekerja dengan menambahkan sampel secara acak dari kelas minoritas hingga mencapai proporsi yang diinginkan dengan kelas mayoritas.

6. RandomUnderSampler
- RandomUnderSampler mengurangi jumlah sampel dari kelas mayoritas secara acak hingga mencapai proporsi yang diinginkan dengan kelas minoritas.
Keuntungan: Mengurangi jumlah sampel mayoritas untuk mencapai keseimbangan dengan kelas minoritas.

7. SMOTE-ENN (SMOTE Edited Nearest Neighbors)
- SMOTE-ENN menggabungkan SMOTE dengan penghapusan sampel menggunakan algoritma Edited Nearest Neighbors (ENN). Setelah menerapkan SMOTE, ENN digunakan untuk menghapus sampel yang dianggap "bersisian" dengan kelas mayoritas, sehingga meningkatkan kualitas sampel sintetis.

#### tanpa scaling

```python
models = [logreg,knn,dt,rf,abc,xgb,lgbm,gbc]
list_model = []
list_sampling = []
score = []
std = []

for sampler in samplers:
    for model in models:
        model_cv = cross_val_score(estimator = Pipeline([
        ('preprocess', transformer),
        ('balancing', sampler),
        ('model', model)]),
                    X = X_train_val,
                    y = y_train_val,
                    cv = skfold,
                    scoring = 'recall',
                    error_score='raise',
                    n_jobs=-1
        )

        list_sampling.append(sampler)
        list_model.append(model)
        score.append(model_cv.mean())
        std.append(model_cv.std())

pd.DataFrame({'model' : list_model, 'recall_score' : score, 'sdev' : std, 'resampler':list_sampling}).set_index('model').sort_values('recall_score', ascending=False).head(10)
```

## XGBoost

XGBoost (eXtreme Gradient Boosting) adalah sebuah implementasi dari algoritma gradient boosting yang dirancang untuk kecepatan dan kinerja yang tinggi. XGBoost Classifier adalah versi dari XGBoost yang digunakan untuk klasifikasi. Algoritma ini menggunakan struktur data yang disebut pohon keputusan (decision trees) dan mengoptimalkannya melalui proses gradient boosting. Berikut penjelasan ringkas cara kerja XGBClassifier:
1. Initialization: XGBoost dimulai dengan model sederhana, seperti decision tree, yang membuat prediksi dasar dengan tebakan sederhana berdasarkan hasil rata-rata data.
1. Tree Building:  Membangun pohon dengan memisahkan data untuk meminimalkan loss function.
1. Gradient Computation: Menghitung Gradient dari loss function untuk setiap instance untuk memandu pembelajaran.
1. Tree Pruning: Memangkas pohon setelah kedalaman maksimum tercapai atau tidak ada pemisahan lebih lanjut yang akan meningkatkan model.
1. Weight Update: Menyesuaikan bobot model menggunakan penurunan gradien untuk mengurangi meminimalkan loss function.
1. Adding Trees: Memperkenalkan pohon baru untuk mengurangi kesalahan pelatihan secara keseluruhan.
1. Shrinking Learning Rate: Terkadang menurunkan kecepatan pembelajaran agar menyatu dengan lancar menuju solusi terbaik.
1. Regularization: Menerapkan regularisasi untuk menghindari overfitting, yang merupakan keunggulan utama dibandingkan boosting standar.
1. Iteration or Termination: Mengulangi proses hingga sejumlah pohon bertambah atau perbaikan menjadi terbatas.
1. Prediction: Setelah dilatih, model memprediksi hasil pada data baru.

XGBoost menonjol karena kemampuannya untuk menangani data yang berukuran besar, kemampuannya dalam melakukan regularisasi, dan efisiensinya dalam melakukan komputasi.

## NearMiss versi 2

NearMiss versi 2 (NearMiss-2) adalah teknik undersampling yang dirancang untuk mengatasi ketidakseimbangan kelas dalam dataset dengan secara selektif menghapus sampel-sampel dari kelas mayoritas yang terjauh dari instansi kelas minoritas. Algoritma ini menghitung jarak antara setiap sampel dari kelas mayoritas dan instansi kelas minoritas, dengan menekankan untuk mempertahankan sampel-sampel yang beragam dan memiliki perbedaan yang lebih besar dari kelas minoritas. Sampel-sampel dari kelas mayoritas yang terpilih kemudian dihapus, menghasilkan dataset yang lebih seimbang dan dapat meningkatkan kinerja model machine learning dalam mendeteksi kelas minoritas. Meskipun NearMiss-2 bertujuan untuk meningkatkan recall untuk kelas minoritas, hal ini dapat menyebabkan penurunan akurasi dan presisi secara keseluruhan karena penghapusan sampel-sampel yang mungkin informatif dari kelas mayoritas. Pemilihan varian yang tepat, termasuk NearMiss-2, tergantung pada karakteristik dataset dan pertimbangan trade-off yang diinginkan antara metrik evaluasi yang berbeda.

```python
resampler = NearMiss(version=2)

model = XGBClassifier(random_state=2023)

estimator=Pipeline([
        ('Preprocess', transformer),
        ('scaler', scaler),
        ('Balancing', resampler),
        ('model', model)
])
```

```python
#xgb
hyperparam_space_xgb = {
    'model__max_depth': [3, 4, 5],
    'model__learning_rate': [0.01, 0.1, 0.2],
    'model__n_estimators': [100, 150, 200],
    'model__subsample': [0.8, 0.9, 1.0],
    'model__min_child_weight': [1, 2, 4],
    'model__gamma': [0, 0.1, 0.2],
    'model__random_state': [42]
}
```

Hyperparameter tuning adalah proses mencari kombinasi optimal dari nilai hyperparameter untuk meningkatkan kinerja model. Dalam konteks tuning hyperparameter XGBoost:

1. 'model__max_depth': Menentukan kedalaman maksimum dari setiap pohon keputusan dalam model. Nilai yang lebih besar dapat meningkatkan kapasitas model, tetapi berpotensi menyebabkan overfitting. Pilihan nilai: 3, 4, 5.

2. 'model__learning_rate': Merupakan laju pembelajaran yang mengontrol kontribusi dari setiap pohon keputusan terhadap model. Nilai yang lebih kecil menyebabkan pembelajaran yang lebih lambat tetapi lebih akurat. Pilihan nilai: 0.01, 0.1, 0.2.

3. 'model__n_estimators': Menentukan jumlah pohon keputusan yang akan dibuat dalam model. Pilihan nilai: 100, 150, 200.

4. 'model__subsample': Menentukan fraksi sampel yang akan digunakan dalam setiap boosting round. Nilai yang lebih kecil dapat membantu mencegah overfitting. Pilihan nilai: 0.8, 0.9, 1.0.

5. 'model__min_child_weight': Menentukan jumlah minimum sampel yang dibutuhkan di dalam setiap leaf node. Nilai yang lebih besar dapat membantu mencegah overfitting. Pilihan nilai: 1, 2, 4.

6. 'model__gamma': Menentukan nilai minimum fungsi objektif yang harus dicapai sebelum membuat pemisahan. Nilai yang lebih besar dapat menghasilkan pemisahan yang lebih konservatif. Pilihan nilai: 0, 0.1, 0.2.

7. 'model__random_state': Seed acak yang digunakan untuk memastikan reproduktibilitas hasil. Tetapkan ke nilai tertentu agar eksperimen dapat direproduksi.

Proses hyperparameter tuning melibatkan mencoba kombinasi nilai-nilai ini untuk mengidentifikasi kombinasi yang memberikan kinerja terbaik pada data validasi atau data uji.

```python
grid = GridSearchCV(estimator, n_jobs=-1, param_grid=hyperparam_space_xgb, scoring='recall',
                    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=2023))
```
